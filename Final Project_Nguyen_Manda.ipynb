```python
#Regression Model
#Less restrictive
import statsmodels.api as sm

X = final_df[["danceability","energy","tempo"]]
y = final_df["popularity"]

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

print(model.summary())
```

```python
#Multicollinearity, there were large condition numbers
#There seems to be no correlation between the 3 
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif["Variable"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

vif
```

```python
#Research Question 3: Can the audio features predict if a song will be highly popular?
final_df["popular"] = (final_df["popularity"] >= 70).astype(int)
import statsmodels.api as sm

X = final_df[["danceability","energy","tempo"]]
y = final_df["popular"]

X = sm.add_constant(X)

logit_model = sm.Logit(y, X).fit()

print(logit_model.summary())
```

```python
#The p-value is 6.094e-122 is less .05 which means there is statistically significance between these
#audio features and its popularity. After looking into the other aspect of the model, the pseudo r-swquared is 0.01205,
#which represents a small proportion in a song's popularity which mean other audio aspects can 
#be a larger contributor to the song's popularity.
```
